### Context

League of Legends adalah MOBA (Multiplayer Online Battle Arena) di mana dalam permainannya terdapat 2 tim (biru dan merah) yang saling berhadapan. Dalam satu laga permainan, ada 3 jalur yang dijadikan medan perang, hutan dan pemainnya terbagi jadi 5 peran. Tujuan permainan ini adalah menjatuhkan Nexus atau base musuh dan memenangkan permainan.

### Content

Diketahui dataset berisi 10 menit dari pertandingan.  Game yang dimainkan berperingkat 10k (SOLO QUEUE Player) dari ELO Diamond I hingga level Master. ELO sendiri merupakan teknik pencarian pemain dengan peringkat atau tingkat keterampilan yang sejenis. Setiap permainan yang terjadi itu unik. GameId dapat membantu kamu mengambil lebih banyak atribut dari Riot API pada kolom dataset. 

Dari data Glossary & Variables, diketahui terdapat 19 fitur pertim (total 38) yang dikumpulkan setelah 10 menit game berlangsung. Data ini termasuk pembunuhan, kematian, emas, pengalaman dan level. Di sini kamu dibebaskan untuk melakukan beberapa rekayasa fitur untuk mendapatkan lebih banyak wawasan atau insight. 

Pada data diketahui juga bahwa kolom blueWins adalah nilai target atau nilai yang akan coba diprediksi, di mana jika Nilai 1 berarti tim biru menang, sedangkan jika Nilai 0 maka sebaliknya. 

### Glossary

- Warding totem: Item yang dapat diletakkan pemain di peta untuk mengungkap area terdekat. Sangat berguna untuk kontrol peta/tujuan. 
- Minions: Item yang dapat diletakkan pemain di peta untuk mengungkap area terdekat. Sangat berguna untuk kontrol peta/tujuan.
- Jungle minions: NPC milik NO TEAM. Mereka memberi emas dan buff saat dibunuh oleh pemain.
- Elite monsters: Monster dengan hp/damage tinggi yang memberikan bonus besar (gold/XP/stat) saat dibunuh oleh tim.
- Dragons: Monster elit yang memberikan bonus tim saat terbunuh. Naga ke-4 yang dibunuh oleh tim memberikan bonus statistik yang sangat besar. The - - 5th dragon (Elder Dragon) menawarkan keuntungan besar bagi tim.
- Herald: Monster elit yang memberikan bonus statistik saat dibunuh oleh pemain. Ini membantu untuk mendorong jalur dan menghancurkan struktur.
- Towers: Struktur yang harus Anda hancurkan untuk mencapai Nexus musuh. Mereka memberi emas.
- Level: Champion level. Mulai dari 1. Maksimal 18.

#### Importing required libraries

```python
from sklearn.preprocessing import MinMaxScaler
```

### Scaling Data

Scalling disini menggunakan konfigurasi minmaxscaler sehingga nilai akan berada pada rentang [0,1]

```python
kolom = data.columns
dataScaling = data

for namaKolom in kolom:
    scaler = MinMaxScaler()
    dataFitTransform = np.array(dataScaling[namaKolom]).reshape(-1,1)
    scaler.fit(dataFitTransform)
    dataTransform = scaler.transform(dataFitTransform)
    dataScaling[namaKolom] = dataTransform

dataScaling
```

data dipecah menjadi y dan x,
y = kolom bluewin
x = kolom data variabilitas

```python
y = dataScaling["blueWins"].values
x = dataScaling.drop(["blueWins"], axis=1).values
```

## 3. Feature Selection

### Principal Component Analysis

PCA akan mereduksi dimensi yang semula berjumlah 30 menjadi 11,
diambil 11 karena ROT dari data pada umumnya berada pada >95%,

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=11)
pca.fit(x, y)
X_PCA = pca.transform(x)
print(np.cumsum(pca.explained_variance_ratio_))
```

## 4. Data Splitting into train and test set

data split menjadi train dan test

```python
from sklearn.model_selection import train_test_split

X_Train, X_Test, Y_Train, Y_Test = train_test_split(X_PCA, y, test_size=0.30, random_state=42)
```

## 5. Modelling and Compare Each Model

### Penyimpanan nilai akurasi hasil pemodelan

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
num_estimators=100
rf = RandomForestClassifier(n_estimators=num_estimators) 
rf.fit(X_Train, Y_Train)

rf_prediksi = rf.predict(X_Test)
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=20)
KNN.fit(X_Train, Y_Train)

KNN_prediksi = KNN.predict(X_Test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(random_state=0)
LR.fit(X_Train, Y_Train)

LR_prediksi = LR.predict(X_Test)
```

Decission Tree Classification

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier(random_state=0)
DTC.fit(X_Train, Y_Train)

DTC_prediksi = DTC.predict(X_Test)
```

Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

GNB = GaussianNB()
GNB.fit(X_Train, Y_Train)

GNB_prediksi = GNB.predict(X_Test)
```

Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
GBC = GradientBoostingClassifier(n_estimators=200, learning_rate=0.1, max_depth=1, random_state=0)
GBC.fit(X_Train, Y_Train)

GBC_prediksi = GBC.predict(X_Test)
```

Stochastic Gradient Descent

```python
from sklearn.linear_model import SGDClassifier

SGD = SGDClassifier(loss="hinge", penalty="l2")
SGD.fit(X_Train, Y_Train)
SGD_prediksi = SGD.predict(X_Test)
```

Support Vector Machines

```python
from sklearn import svm

SVM = svm.SVC()
SVM.fit(X_Train, Y_Train)
SVM_prediksi = SVM.predict(X_Test)
```

Gaussian Process Classification

```python
from sklearn.gaussian_process import GaussianProcessClassifier

GPC = GaussianProcessClassifier(random_state=0)
GPC.fit(X_Train, Y_Train)
GPC_prediksi = GPC.predict(X_Test)
```

## 6. Modelling and Evaluation


```python
from sklearn.metrics import precision_score, \
    recall_score, confusion_matrix, classification_report, \
    accuracy_score, f1_score, ConfusionMatrixDisplay, roc_curve , roc_auc_score

def model_evaluation(model,name):
    confusion_matrix = pd.crosstab(Y_Test, model, rownames=['Actual'], colnames=['Predicted'], margins = True)
    sns.heatmap(confusion_matrix, square=True, annot=True, fmt='d', cbar=False)
    plt.xlabel('Prediction label')
    plt.ylabel('True Label')
    plt.title(name)
    plt.yticks([0.5,1.5], [ 'False', 'True'],va='center')
    plt.xticks([0.5,1.5], [ 'False', 'True'],va='center')
    plt.show()

# the function that prints all scores
def print_scores(headline, y_pred, y_true=Y_Test):
    print("==========",headline,"==========")
    acc_score = accuracy_score(y_true, y_pred)
    print("accuracy: ",acc_score)
    pre_score = precision_score(y_true, y_pred)
    print("precision: ",pre_score)
    rec_score = recall_score(y_true, y_pred)                            
    print("recall: ",rec_score)
    f_score = f1_score(y_true, y_pred, average='weighted')
    print("f1_score: ",f_score)
    print("\n")
    return headline, acc_score, pre_score, rec_score, f_score
```

```python
hasil = []
```

## Melihat Nilai dari akurasi, presisi, recall, dan f_1 Score dari setiap pengujian

```python
ml = [
    ["Random Forest Classifier", rf_prediksi],
    ["KNN", KNN_prediksi],
    ["Logistic Regression", LR_prediksi],
    ["Decission Tree Classification", DTC_prediksi],
    ["Gaussian Naive Bayes", GNB_prediksi],
    ["Gradient Boosting Classifier", GBC_prediksi],
    ["Stochastic Gradient Descent", SGD_prediksi],
    ["Support Vector Machines", SVM_prediksi],
    ["Gaussian Process Classification", GPC_prediksi],
]
```

```python
for i in range(len(ml)):
    hasil.append(print_scores(headline=ml[i][0], y_pred=ml[i][1]))
```

```python
df_hasil = pd.DataFrame(hasil, columns =['Names', 'Accuracy', 'Precission', 'Recall', 'F1_Score'])

df_hasil.sort_values('Accuracy', ascending=False)
```

#### Logistic Regression, Gradient Boosting Classifier, dan Gaussian Process Classification
#### memiliki nilai top 3 f1_score, sehingga diakhir akan ditampilkan dari 3 model tersebut

```python
model_evaluation(GPC_prediksi, "Gaussian Process Classification")
model_evaluation(GBC_prediksi, "Gradient Boosting Classifier ")
model_evaluation(LR_prediksi, "Logistic Regression")
```

#### Kesimpulan

Logistic Regression, Gradient Boosting Classifier, dan Gaussian Process Classification adalah top 3 nilai f1_score dari hasil pemodelan ini, 3 tipe regresi ini merupakan tipe pemodelan acak yang dapat dimanfaatkan untuk variabel yang banyak. Memang masih perlu pendalaman lebih lanjut seperti penggunaan Deep Learning, karena dengan penggunaan klasifikasi ini masih memiliki nilai akurasi diangka 70%. Mengapa ini dapat terjadi? Ini dapat terjadi karena data begitu multivariatif, permainan tidak hanya berbicara angka, namun juga GAMEPLAY dari pemain itu sendiri, sehingga sulit untuk memprediksi dengan angka saja.
