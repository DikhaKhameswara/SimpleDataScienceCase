#### Read CSV File

```python
train = pd.read_csv("train.csv")
smartphonePrediction = pd.read_csv("test.csv")
smartphonePrediction = smartphonePrediction.drop("id", axis=1)
```

#### Show Column Name

```python
kolom = train.columns
kolom = kolom.delete([1, 3, 5, 17, 18, 19, 20])
kolom
```

#### Scaling Data

```python
for namaKolom in kolom:
    data = np.array(train[namaKolom]).reshape(-1,1)
    scaler = MinMaxScaler()
    scaler.fit(data)
    scaler.transform(data)
    train[namaKolom] = scaler.transform(data)

for namaKolom in kolom:
    data = np.array(smartphonePrediction[namaKolom]).reshape(-1,1)
    scaler = MinMaxScaler()
    scaler.fit(data)
    scaler.transform(data)
    smartphonePrediction[namaKolom] = scaler.transform(data)
```

```python
train.head()
```

```python
smartphonePrediction.head()
```

#### Spliting data to X and Y array

```python
Y = np.array(train["price_range"])
X = np.array(train.drop("price_range", axis=1))
```

#### Splitting data to Train and Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

## Create Deep Learning Model using ANN

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='sigmoid'),
    tf.keras.layers.Dense(84, activation='sigmoid'),
    tf.keras.layers.Dense(63, activation='sigmoid'),
    tf.keras.layers.Dense(21, activation='sigmoid'),
    tf.keras.layers.Dense(4,  activation='softmax')
])
model.compile(optimizer='Adam',
              loss='sparse_categorical_crossentropy')
```

## Fitting Data Train

```python
model.fit(X_train, y_train, epochs=150, batch_size=32)
```

```python
test_loss = model.evaluate(X_test,  y_test, verbose=2)

print('\nTest Loss: {0:2.2f}%'.format(test_loss*100.0))
```

## Predict Value

```python
predictions = model.predict(X_test)
```

## Show Error Value from Y_Test

```python
error = 0
for i in range(len(y_test)):
    if y_test[i] != np.argmax(predictions[i]):
        error += 1

print("error percentage={0:2.2f}%".format((error/len(y_test))*100.0))
```

## Predicting Price for Smartphone

```python
SmartphonePriceForPredict = np.array(smartphonePrediction)
```

```python
predictSmartphonePrice = model.predict(SmartphonePriceForPredict) * 100
```

## Show value testing

```python
predictSmartphonePrice[0]
```

```python
mainPrice = []
secPrice = []
for i in range(len(SmartphonePriceForPredict)):
    mainPrice.append(np.argmax(predictSmartphonePrice[i]))
    predictSmartphonePrice[i][np.argmax(predictSmartphonePrice[i])] = 0
    secPrice.append(np.argmax(predictSmartphonePrice[i]))
```

## Create DataFrame for Price Prediction

```python
rangePrice = pd.DataFrame(data={'id': np.arange(1,1001), 'Main Price': mainPrice, 'Seccond Price': secPrice})
rangePrice
```
